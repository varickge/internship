```python
def largest_same_digit_image(grid):
    max_area = 0
    rows = len(grid)
    cols = len(grid[0])

    def is_valid(x, y, digit, grid, visited):
        if x >= 0 and x < rows and y >= 0 and y < cols and grid[x][y] == digit and not visited[x][y]:
            return True
        return False

    def dfs(x, y, digit, grid, visited):
        dx = [-1, 1, 0, 0]
        dy = [0, 0, -1, 1]
        visited[x][y] = True
        area = 1
        for i in range(4):
            x_new = x + dx[i]
            y_new = y + dy[i]
            if is_valid(x_new, y_new, digit, grid, visited):
                area += dfs(x_new, y_new, digit, grid, visited)
        return area

    visited = [[False for _ in range(cols)] for _ in range(rows)]
    for i in range(rows):
        for j in range(cols):
            if str(grid[i][j]).isdigit() and not visited[i][j]:
                area = dfs(i, j, grid[i][j], grid, visited)
                max_area = max(max_area, area)
    return max_area

array =[[1, 3, 2, 3, 3, 3],
        [3, 3, 3, 2, 4, 4],
        [4, 3, 1, 2, 3, 3],
        [4, 3, 1, 3, 3, 1],
        [4, 3, 3, 3, 1, 1] ]

print(largest_same_digit_image(array))
```
